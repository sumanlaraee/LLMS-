```python
import h2o
from h2o.automl import H2OAutoML

h2o.init()
```

```python
data = h2o.import_file("/content/synthetic_fraud_detection_data_2000_labeled.csv")
```

```python
train, test = data.split_frame(ratios=[0.8], seed=1234)
```

```python
aml = H2OAutoML(max_runtime_secs=600, seed=1)
aml.train(y='is_fraud', training_frame=train)
```

```python
lb = aml.leaderboard
aml.leader.model_performance(test_data=test)
```
